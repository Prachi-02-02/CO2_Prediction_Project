```python
#Ml tutorial 
#mean is the average value
import numpy
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
mean = numpy.mean(speed)
print(mean) #89.76
```

```python
#meadian is the value in the middle after all values are sorted
import numpy
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
median = numpy.median(speed)
print(median) # 87.0
```

```python
#if there are two numbers in the middle divide the sum by two numbers
import numpy
speed = [99,86,87,88,86,103,87,94,78,77,85,86]
median1 = numpy.median(speed)
print(median1) #86.5
```

```python
#mean value is the value that appears most of the time 
from scipy import stats
speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
mode = stats.mode(speed)
print(mode)  # 86 (3)
```

```python
# std deviation is the number  which tells how spread out the  values are
import numpy
speed = [86,87,88,86,87,85,86]
deviation = numpy.std(speed)
print(deviation) #0.9 low deviation values are closer to the mean value 
mean = numpy.mean(speed)
print(mean)
```

```python
import numpy
speed = [32,111,138,28,59,77,97]
deviation1= numpy.std(speed)
print(deviation1) #37.84 means high deviation values are spread over wider range 
```

```python
#variance is also a number which tells how spred out the values are
#sq root of variance is the std deviation
#multiply std deviation by itself  is the variance
import numpy 
speed = [32,111,138,28,59,77,97]
variance = numpy.var(speed)
print(variance)
```

```python
#linear regression
import matplotlib.pyplot as plt
from scipy import stats
car_age = [5,7,8,7,2,17,2,9,4,11,12,9,6]
car_speed = [99,86,87,88,111,86,103,87,94,78,77,85,86]
slope , intercept , r, p , std_err = stats.linregress(car_age, car_speed)
def myfunc(x):
    return slope * x + intercept
model = list(map(myfunc , x))
plt.scatter(car_age , car_speed)
plt.plot(x, model)
plt.show()
```

```python
#the relationship between the values of the x-axis and the values of the y-axis is, if there are no relationship the linear regression can not be used to predict anything.

#This relationship - the coefficient of correlation - is called r.

#The r value ranges from -1 to 1, where 0 means no relationship, and 1 (and -1) means 100% related.
```

```python
#multi- linear regression 
#similar to linear regression but more than one independent variable means that more than one value can be used to make prediction 
```

```python
import pandas
df = pandas.read_csv(r'C:\Users\ADMIN\Machine Learning\data.csv')
```

```python
# It is common to name the list of independent values with a upper case X, and the list of dependent values with a lower case y.
X = df[['weight','volume']] # independant  variables name it as X 
#When selecting multiple columns, you need to use double brackets ([[...]]) instead of a comma inside a single bracket: unless it will give keyerror
y = df['CO2'] #name dependent variable as Y
```

```python
#import sklearn module as we have to use methods in it
from sklearn import linear_model
regres = linear_model.LinearRegression() #use lineaRegressson method  to create a linear object
regres.fit(X,y) #fills the linear object with the data that describes relation
```

```python
#now we have a regression object that is ready to predict the C02 value based on car's volume and weight
```

```python
import pandas as pd

predictC02 = regres.predict(pd.DataFrame([[2300, 1300]], columns=['weight', 'volume'])) #Convert Input to a DataFrame ,it is best practice to pass data in the same format as used for training.
print(predictC02) #We have predicted that a car with 1.3 liter engine, and a weight of 2300 kg, will release approximately 107 grams of CO2 for every kilometer it drives.
```

```python
#predicting the CO2 emmission of a car with 1.6 liter engine and a weigght of 1252 kg
import pandas as pd
df = pd.read_csv(r'C:\Users\ADMIN\Machine Learning\data.csv')
X = df[['weight','volume']]
y= df['CO2']
from sklearn import linear_model
regres = linear_model.LinearRegression()
regres.fit(X,y)
predictCO2 = regres.predict(pd.DataFrame([[1252, 1600]] ,  columns = ['weight','volume']))
print(predictCO2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate weight values for plotting
weight_range = np.linspace(df['weight'].min(), df['weight'].max(), 100)
volume_avg = df['volume'].mean()  # Use mean volume for fixed value

# Predict CO2 emissions for different weights
predicted_CO2 = regres.predict(np.column_stack((weight_range, np.full_like(weight_range, volume_avg))))

# Scatter plot of actual data
plt.scatter(df['weight'], y, color='blue', label="Actual Data")

# Regression line
plt.plot(weight_range, predicted_CO2, color='red', label="Regression Line")

# Highlight the new prediction
plt.scatter(1252, predictCO2, color='green', s=100, label="Predicted CO₂")

# Labels & legend
plt.xlabel("Weight (kg)")
plt.ylabel("CO2 Emission")
plt.title("CO2 Emission vs Weight (Volume Fixed)")
plt.legend()
plt.show()

```
